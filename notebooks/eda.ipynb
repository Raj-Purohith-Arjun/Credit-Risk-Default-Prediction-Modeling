# Credit Risk EDA

```python
import sys
sys.path.insert(0, '..')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from src.data.data_loader import load_raw
from src.data.preprocess import preprocess
from src.features.feature_engineering import engineer_features

sns.set_theme(style='whitegrid')
%matplotlib inline
```

```python
df = load_raw()
X, y = preprocess(df, save=False)
X = engineer_features(X)
data = X.copy()
data['default'] = y
print(data.shape)
data.head()
```

```python
# target distribution
ax = data['default'].value_counts().plot(kind='bar', rot=0)
ax.set_title('Default Distribution')
ax.set_xlabel('Default')
ax.set_ylabel('Count')
plt.tight_layout()
```

```python
# numeric distributions
num_cols = ['loan_amnt', 'annual_inc', 'dti', 'fico_range_low', 'revol_util']
fig, axes = plt.subplots(1, len(num_cols), figsize=(18, 4))
for ax, col in zip(axes, num_cols):
    data[col].hist(bins=40, ax=ax)
    ax.set_title(col)
plt.tight_layout()
```

```python
# default rate by fico band
data.groupby('fico_band')['default'].mean().sort_values().plot(kind='bar', rot=30)
plt.title('Default Rate by FICO Band')
plt.ylabel('Default Rate')
plt.tight_layout()
```

```python
# correlation heatmap
numeric_data = data.select_dtypes(include='number')
corr = numeric_data.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
fig, ax = plt.subplots(figsize=(12, 10))
sns.heatmap(corr, mask=mask, annot=False, cmap='RdBu_r', center=0, ax=ax)
ax.set_title('Feature Correlations')
plt.tight_layout()
```

```python
# loan-to-income vs default
fig, ax = plt.subplots(figsize=(7, 4))
data.boxplot(column='loan_to_income', by='default', ax=ax)
ax.set_title('Loan-to-Income by Default Status')
ax.set_xlabel('Default')
plt.suptitle('')
plt.tight_layout()
```
